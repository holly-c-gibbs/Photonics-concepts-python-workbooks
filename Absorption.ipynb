Complex representation of waves provides a nice mathematical framework for describing a material's ability to absorb the energy of waves. The way this absorption is described is with a complex-valued material property that impacts the wavenumber, $k$. If $k$ is complex, by the properties of exponentials, **there will be a seperable component of the exponential that is real. This real exponential term typically is negative and thus dampens the wave.**

In a dielectric material, the wavenumber $k$ can be related to the material's electric susceptilibty:

\begin{equation}
k = k_0 \sqrt{1 + \chi}
\end{equation}

If $\chi$ is complex-valued, then:

\begin{equation}
k = k_0 \sqrt{1 +\chi ' + j\chi ''}
\end{equation}

where $\chi '$ is the real component and $\chi ''$ is the imaginary component of thte complex-valued electric susceptibility, and $k_0 = \omega / c_0$ is the wavenumber in free space.

Let's notice that there is a real and imaginary part of $k$ and decide to seperate them to make life easier. If we choose $\alpha$ to represent the imaginary part and $\beta$ to represent the real part of $k$, we can write:

\begin{equation}
k = \beta - \frac{j\alpha}{2}
\end{equation}

Then if we describe a monochromatic plane wave's complex amplitude:

\begin{equation}
U(z) = A_0 exp(-jkz) = A_0exp(-j(\beta- \frac{j\alpha}{2})z)
\end{equation}

which becomes:
\begin{equation}
U(z) = A_0 exp(-jkz) = A_0exp(- \frac{\alpha z}{2})exp(-j\beta z)
\end{equation}

Notice $\beta$ acts to control the rate at which the phase of the wave changes as we saw with $k$ in free space. For this reason, $\beta$ is called the **propagation constant**. The new part, $\alpha$, is called the **absorption coefficient** and decreases the intensity of the wave according to $exp(-\alpha z)$.

Different assumptions can be made for weakly v. strongly absorbing media, providing different expressions for $\alpha$ and $\beta$.
