**Exercise 1** Gaussian functions are used in many applications in statistics and physics. Let's remind ourselves of the form of this function and how the important parameters affect the function's shape. A 1-dimensional Gaussian function can be described as:

\begin{equation}
G(x)=ae^{-(x-b)^2/2c^2}
\end{equation} 

Plot this equation and see what happens when you vary the parameters a, b, and c.

```python
import numpy as np
import matplotlib.pyplot as plt

#constants
a_array=[1,2,3] #controls height of peak
b_array=[1,2,3] #controls center position of peak, like mean of a dataset
c_array=[1,2,3] #controls width of peak, like standard deviation of a normal dataset
a=1
b=1
c=1

#spatial array
x=np.linspace(-10,10,500)

#Gaussian with different a values
g1=a_array[0]*np.exp(-np.square(x-b)/(2*np.square(c)))
g2=a_array[1]*np.exp(-np.square(x-b)/(2*np.square(c)))
g3=a_array[2]*np.exp(-np.square(x-b)/(2*np.square(c)))

#plot
plt.figure(1)
plt.plot(x,g1,x,g2,x,g3)
plt.xlabel('x')
plt.ylabel('G(x)')
plt.legend(['a=1','a=2','a=3'])
plt.title('a controls the peak height')

#Gaussian with different b values
g4=a*np.exp(-np.square(x-b_array[0])/(2*np.square(c)))
g5=a*np.exp(-np.square(x-b_array[1])/(2*np.square(c)))
g6=a*np.exp(-np.square(x-b_array[2])/(2*np.square(c)))

#plot
plt.figure(2)
plt.plot(x,g4,x,g5,x,g6)
plt.xlabel('x')
plt.ylabel('G(x)')
plt.legend(['b=1','b=2','b=3'])
plt.title('b controls the peak position in x')

#Gaussian with different c values
g7=a*np.exp(-np.square(x-b)/(2*np.square(c_array[0])))
g8=a*np.exp(-np.square(x-b)/(2*np.square(c_array[1])))
g9=a*np.exp(-np.square(x-b)/(2*np.square(c_array[2])))

#plot
plt.figure(3)
plt.plot(x,g7,x,g8,x,g9)
plt.xlabel('x')
plt.ylabel('G(x)')
plt.legend(['c=1','c=2','c=3'])
plt.title('c controls the peak width')
```

**Exercise 2:** Let's try a two-dimensional Gaussian function now, using the expression: 

\begin{equation}
G(r)=ae^{-r^2/2c^2}
\end{equation} 

where $r^2=x^2+y^2$. Using numpy's meshgrid function can be a useful way to calculate a function in 2D. The meshgrid function takes two 1-dimensional arrays, for example, and x array and y array, and outputs 2-dimensional grids whose values are coordinates, in this case x and y coordinates. Then you can use simple multiplication to calculate the function. Take a line plot to see how it compares to the 1-dimensional Gaussian.

```python
#second spatial array
y=np.linspace(-10,10,500)
xx,yy=np.meshgrid(x,y)

#visualize x,y grids
plt.figure(1)
plt.imshow(xx)
plt.title('x grid coordinates')
plt.figure(2)
plt.imshow(yy) #notice y coordinate axis is flipped, with 0 on the top
plt.title('y grid coordinates')

#create and visualize r grid
r=np.sqrt(np.square(xx)+np.square(yy))
plt.figure(3)
plt.imshow(r)
plt.title('r values for given x,y')

#calculate 2-dimensional Gaussian
g2d=a*np.exp(-np.square(r)/(2*np.square(c)))
plt.figure(4)
plt.imshow(g2d)
plt.title('2-dimensional Gaussian')

#line plot through maximum
plt.figure(5)
plt.plot(x,g2d[:,250])
plt.title('Line plot through center of radial 2D Gaussian')
plt.xlabel('x')
plt.ylabel('G(r),y=0')
```

**Exercise 3** Next let's examine the beam waist radius, $W(z)$ along the optical axis as defined for a Gaussian beam that satisfies the wave equation.

\begin{equation}
W(z)=W_0\sqrt{1+\left( \frac{z}{z_0} \right)^2}
\end{equation}

Plot this for different size beams. You will need to choose a values of $W_0$, which you can use to calculate $z_0$ for a given wavelength $\lambda$ using $z_0=\pi W_0^2/ \lambda$. Recall $z_0$ is the designated **Rayleigh range** which **is half the depth of focus** of the beam.

```python
#constants
W_01=0.5e-6 #beam waist 0.5 micron
W_02=1e-6
W_03=1.5e-6
lam=488e-9 
z_01=np.pi*np.square(W_01)/lam
z_02=np.pi*np.square(W_02)/lam
z_03=np.pi*np.square(W_03)/lam

#optical axis
z=np.linspace(-20e-6,20e-6,1000)

#calculating beam waist function
Wz1=W_01*np.sqrt(1+np.square(z/z_01))
Wz2=W_02*np.sqrt(1+np.square(z/z_02))
Wz3=W_03*np.sqrt(1+np.square(z/z_03))

#plot
plt.plot(z,Wz1,z,Wz2,z,Wz3)
plt.legend(['W_0=half micron','W_0=1 micron','W_0=1.5 microns'])
plt.xlabel()
```

**Exercise 4** Now let's plot the intensity of a Gaussian beam, which is a 3-dimensional function. If we use a cylindrical coordinate system, we have $\rho$ as the radial dimension. Recall $\rho^2 = x^2 + y^2$. The third dimension is the optical axis, $z$. The expression for the beam's intensity is: 

\begin{equation}
I(\rho,z)=I_0 \left[ \frac{ W_0 } {W(z)}  \right]^2exp \left[ \frac{-2 \rho^2} {W^2(z)}\right] 
\end{equation} 

Notice how $W(z)$, the beam waist changes the peak intensity as a function of z, which is like the $a$ term in our example Gaussian. As $W(z)$ gets bigger, the peak intensity of the Gaussian beam decreases. Notice also how $W(z)$ changes the width of the Gaussian like we saw with the $c$ parameter, as $W(z)$ gets bigger, the width of the Gaussian increases. Explore this change in the Gaussian beam as a function of $z$. 

Take $I_0=1$. Be sure to plot $I(\rho,0)$, $I(\rho,z_0)$, and $I(\rho,2 z_0)$ to find what is significant about $z_0$. How does the power of the beam change as a function of z?

**Hint:** If use a separate cell to do the visualizations so you don't have to keep re-calculating the Gaussian beam intensity simulation, which could take a while depending on the size of your grids. 

```python
#constants
I_0=1
W_0=1e-6 #beam waist 0.5 micron
lam=488e-9 
z_0=np.pi*np.square(W_0)/lam

#arrays
x=np.linspace(-5e-6,5e-6,100)
y=np.linspace(-5e-6,5e-6,100)
z=np.linspace(-15e-6,15e-6,1000)

#grid
xx,yy,zz=np.meshgrid(x,y,z)
rho_sq=np.square(xx)+np.square(yy)

#Gaussian beam intensity
W_z=W_0*np.sqrt(1+np.square(zz/z_0))
g_int=I_0*np.square(W_0/W_z)*np.exp(-2*rho_sq/np.square(W_z))

#check dimensions that g_int matches grid size
print(np.shape(g_int))
```

```python
#check Raleigh range, make sure grid covers +/- 2*z0
print(z_0)
```

```python
#Explore G(z), note x=0,y=0 is close to indices 50 on the grid

#plot Intensity on the optical axis (x=0,y=0), notice inverse relationship to beam waist (Exercise 3)
plt.figure(1)
plt.plot(z,g_int[50,50,:])
plt.title('Intensity on the optical axis (w/mm^2)')
plt.xlabel('z')
plt.ylabel('Intensity (W/mm^2')

#Show Intensity in an xz plane on the optical axis (y=0) 
plt.figure(2)
plt.subplots(figsize=[30,30])
plt.imshow(g_int[:,50,:],cmap='gray')
plt.title('Intensity on the optical axis, x=0 (W/mm^2)',fontsize=18)
plt.xlabel('z',fontsize=18)
plt.ylabel('x',fontsize=18)
```

```python
#Transverse intensity
#z=0
plt.imshow(g_int[:,:,500],cmap='gray',vmin=0,vmax=1) #note index 500 is closest to z=0
plt.title('I(rho,z=0)')
#peak intensity (notice this changes as a function of z)
print(np.max(np.max(g_int[:,:,500])))
#total power (notice this does not change as a function of z)
print(np.sum(np.sum(g_int[:,:,500])))
```

```python
#z=z_0, the end of the depth of focus, where the intensity has dropped by half
#find correct index to address 
array1=z>z_0
ind1=np.min(np.where(array1==1))
plt.imshow(g_int[:,:,ind1],cmap='gray',vmin=0,vmax=1)
plt.title('I(rho,z=z_0)')
#peak intensity (notice this changes as a function of z)
print(np.max(np.max(g_int[:,:,ind1])))
#total power (notice this does not change as a function of z)
print(np.sum(np.sum(g_int[:,:,ind1])))
```

```python
#z=2z_0
array2=z>2*z_0
ind2=np.min(np.where(array2==1))
plt.imshow(g_int[:,:,ind2],cmap='gray',vmin=0,vmax=1)
plt.title('I(rho,z=2*z_0)')
#peak intensity (notice this changes as a function of z)
print(np.max(np.max(g_int[:,:,ind2])))
#total power (notice this does not change as a function of z)
print(np.sum(np.sum(g_int[:,:,ind2])))
```

```python
#plot line profiles of intensity as function of z
plt.plot(x,g_int[:,50,500],x,g_int[:,50,ind1],x,g_int[:,50,ind2])
plt.xlabel('Distance (m)')
plt.ylabel('Gaussian beam intensity (Power/area)')
plt.legend(['z=0','z=z_0','z=2*z_0'])
```
