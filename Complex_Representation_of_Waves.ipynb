**Exercise 1:** Plot the expression below for the wave as a function of time for a given point in space. You can just use $A=1$ for illustrative purposes here, but go ahead and select realistic numbers for the other variables.

\begin{equation}
u(z,t)=Acos(kz-\omega t)
\end{equation} 

```python
import numpy as np
import matplotlib.pyplot as plt

#some constants
c=3e8; #speed of light 
lam=500e-9; #pick 500 nm for the wavelength, lambda
om=2*np.pi*c/lam #angular frequency
k=2*np.pi/lam #wave number
A=1; 

#relevant length scale, nanometers
z=np.linspace(0.01e-9,5000e-9,1000)
#relevant time scale, femtoseconds 
t=np.linspace(0,10e-15,1000)

#function of time for a given space, take z=0.01e-9
time=A*np.cos(k*0.01e-9-om*t)
plt.plot(z,time)
plt.xlabel('Time (s)')
plt.ylabel('Wave Amplitude')
plt.title('Wave in time for a given space')
```

**Exercise 2:** Plot the expression from Exercise 1 as a function of space for a given time. Pick successive time points and overlay the plots to see how the wave moves.

```python
#function of space for a given time, take t=0
space1=A*np.cos(k*z-om*t[0])
space2=A*np.cos(k*z-om*t[10])
space3=A*np.cos(k*z-om*t[20])
plt.plot(z,space1,z,space2,z,space3)
plt.xlabel('Distance (m)')
plt.ylabel('Wave Amplitude')
plt.title('Wave in space for a given time')
plt.legend(['t1','t2','t3'])
```

**Exercise 3:** Time to get complex! Let's plot the same wave but using complex exponential representation. This approach uses the Euler relation to express a complex exponential as the sum of real and imaginary sinusoidal functions.

\begin{equation}
Ae^{j\theta}=A[cos(\theta)+jsin(\theta)]
\end{equation} 

Now if we use the argument $[kz-\omega t]$ in this expression, we get:

\begin{equation}
Ae^{j(kz-\omega t)}=A[cos(kz-\omega t)+jsin(kz-\omega t)]
\end{equation} 

Notice that the real part of the right hand of the expression above is the same as in Exercises 2 and 3. Because of the property of exponentials that $e^{a+b}=e^{a}e^{b}$, the complex representation is often split into two exponentials to consider. 

\begin{equation}
Ae^{j(kz-\omega t)}=Ae^{jkz}e^{-j\omega t}
\end{equation} 

The expression $U(z)=Ae^{jkz}$ (or as in the convention in the Saleh text, $Ae^{-jkz}$) is called the **Complex Amplitude**. This particular mathematical expression describes a plane wave, incidentally. So, anyway, plot it and see how it compares to results from Exercise 1 and 2.

```python
k=2*np.pi/lam
U_plane=A*np.exp(-1j*k*z) #Use Saleh convention for sign of the spatial component
plt.plot(z,np.real(U_plane),z,space1)
plt.xlabel('Distance (m)')
plt.ylabel('Wave Amplitude')
```

**Exercise 4:** Here let's add in the time part, $e^{j\omega t}$ (Saleh sign convention), to see if we can make the wave move like we did before. What happens to the wave if both complex exponentials have the same sign? 

```python
U_plane_t1=A*np.exp(-1j*k*z)*np.exp(1j*om*t[0])
U_plane_t2=A*np.exp(-1j*k*z)*np.exp(1j*om*t[10])
U_plane_t3=A*np.exp(-1j*k*z)*np.exp(1j*om*t[20])
plt.plot(z,U_plane_t1,z,U_plane_t2,z,U_plane_t3)
plt.xlabel('Distance (m)')
plt.ylabel('Wave Amplitude')
plt.title('Wave in space for a given time')
plt.legend(['t1','t2','t3'])
```

**Exercise 5:** OK, and let's not forget the fun of the intensity, given by:

\begin{equation}
I(z)=|U(z)|^2
\end{equation} 

Plot both the plane wave and its intensity, where $|U(z)|^2$ is $U(z)$ multiplied by its complex conjugate $U(z)^*$. Recall you can define the conjugate of an imaginary number by changing the sign of all the $i$ or $j$ terms.

```python
u_plane=A*np.exp(-1j*k*z)
plt.plot(z,np.real(u_plane),z,u_plane*np.conjugate(u_plane))
plt.xlabel('Distance (m)')
plt.ylabel('A.U.') #We'll do a better job with these units when we discuss electromagnetic fields
plt.legend(['Wave Amplitude','Optical Intensity'])
```
